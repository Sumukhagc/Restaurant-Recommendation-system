```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('zomato.csv')
df.head()
```

```python
df.drop(['url','dish_liked','phone'],axis=1,inplace=True)
df.shape
```

```python
df.duplicated().sum()
df=df.drop_duplicates()
df.info()
```

```python
df.isnull().sum()
```

```python
df['cuisines'].dropna(inplace=True)
df.isnull().sum()
```

```python
df.rename(columns = {'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'}, inplace = True)
df['cost']=df['cost'].astype(str)
df['cost']=df['cost'].apply(lambda x:x.replace(',',''))
df['cost']=df['cost'].astype(float)
df['cost'].fillna(df['cost'].median(),inplace=True)
df['cost']
```

```python
df['online_order']=pd.get_dummies(df['online_order'],drop_first=True)
df['book_table']=pd.get_dummies(df['book_table'],drop_first=True)
df.head()

```

```python
df['rate']=df['rate'].str.replace('NEW','0/5')
df['rate']=df['rate'].str.replace('-','0/5')
df['rate']=df['rate'].str.replace('/5','')
df['rate'].astype(float)
```

```python
def impute_nan(df,variable,median):
    
    df[variable]=df[variable]
    ##It will have the random sample to fill the na
    random_sample=df[variable].dropna().sample(df[variable].isnull().sum(),random_state=0)
    ##pandas need to have same index in order to merge the dataset
    random_sample.index=df[df[variable].isnull()].index
    df.loc[df[variable].isnull(),variable]=random_sample
```

```python
median=df['rate'].median()
impute_nan(df,'rate',median)
df[df['name']=='Airlines Hotel']
```

```python
hotels=[df['name'].unique()]
```

```python
df['rate'] = df['rate'].astype(float)
dic=df.groupby('name').rate.agg(['mean']).to_dict()
dic
df['rate']=df['name'].map(dic['mean'])
df['rate']=df['rate'].round(1)
df.head()
```

```python
df['reviews_list']=df['reviews_list'].str.lower()

import string

def remove_puctuation(text):
    return "".join([x for x in text if x not in string.punctuation ])
df['reviews_list']=df['reviews_list'].apply(lambda x:remove_puctuation(x))
```

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')

def remove_stop_words(text):
    return "".join([x for x in str(text).split() if x not in stop_words])

```

```python
df['reviews_list']=df['reviews_list'].apply(lambda x:remove_stop_words(x))
```

```python
df['reviews_list']
```

```python
import re
def remove_urls(text):
    return re.compile(r'https?://\S+|www\.\S+').sub(r'', text)
    
```

```python
df['reviews_list'] = df['reviews_list'].apply(lambda x: remove_urls(x))
```

```python
df.head()
```

```python
df=df.drop(['address','rest_type', 'type', 'menu_item', 'votes'],axis=1)
```

```python
df_percentage = df.sample(frac=0.5)
df_percentage=df_percentage.set_index(df_percentage.name)
hotel_index = pd.Series(df_percentage.index)
```

```python
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0, stop_words='english')
tfidf_matrix =tfidf.fit_transform(df_percentage['reviews_list'])
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def recommended_hotels(hotel_to_search, cosine_similarities = cosine_similarities):
    hotel_recommended_list = []
    index = hotel_index[hotel_index == hotel_to_search].index[0]
    similar_hotels = pd.Series(cosine_similarities[index]).sort_values(ascending=False)
    top_20_hotels = list(similar_hotels.iloc[0:21].index)
    for each in top_20_hotels:
        hotel_recommended_list.append((list(df_percentage.index)[each]))
    recommended=pd.DataFrame(columns=['name','cuisines', 'rate', 'cost']) 
    for each in hotel_recommended_list:
        recommended= recommended.append(pd.DataFrame(df_percentage[['name','cuisines','rate', 'cost']][df_percentage.name == each].sample()))
    recommended = recommended.sort_values(by='cuisines',ascending=False)
    recommended.drop_duplicates(keep=False,inplace=True)
    recommended = recommended.reset_index(drop=True)
    return recommended.head(10)
recommended_hotels('Pizza Hut')
```
